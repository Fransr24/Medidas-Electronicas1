```python
import math as mt

#Frecuencia 10hz
k95 = 1.65
# Con Tc=1/(fbt*10^k)
Fi = 10
#N = fi*Tc*Promedio (Fi frecuencia indicada siempre 10hz)
N01 = Fi*0.1
N1 = Fi*1
N10 = Fi*10

urTbt = 0.3*12*10**-6
urN01 = 1/N01
urN1 = 1/N1
urN10 = 1/N10

uFi01= k95*(urTbt + urN01)*Fi/mt.sqrt(3)
uFi1= k95*(urTbt + urN1)*Fi/mt.sqrt(3)
uFi10= k95*(urTbt + urN10)*Fi/mt.sqrt(3)

print(f'Incertidumbre de Frecuencia a 10hz indicada en Tc = 0.1s: {uFi01}hz\n')
print(f'Incertidumbre de Frecuencia a 10hz indicada en Tc = 1s: {uFi1}hz\n')
print(f'Incertidumbre de Frecuencia a 10hz indicada en Tc = 10s: {uFi10}hz\n')
print(f'-------------------------------------------------------------------------------\n')

#Frec 1khz
Fi = 1000
Fi10 = 999.9  
N01 = Fi*0.1
N1 = Fi*1
N10 = Fi10*10
urN01 = 1/N01
urN1 = 1/N1
urN10 = 1/N10
uFi01= k95*(urTbt + urN01)*Fi/mt.sqrt(3)
uFi1= k95*(urTbt + urN1)*Fi/mt.sqrt(3)
uFi10= k95*(urTbt + urN10)*Fi10/mt.sqrt(3)

print(f'Incertidumbre de Frecuencia a 1Khz indicada en Tc = 0.1s: {uFi01}hz\n')
print(f'Incertidumbre de Frecuencia a 1Khz indicada en Tc = 1s: {uFi1}hz\n')
print(f'Incertidumbre de Frecuencia a 1Khz indicada en Tc = 10s: {uFi10}hz\n')
print(f'-------------------------------------------------------------------------------\n')

#Frec 100khz
Fi01 = 100000
Fi1 =  99996
Fi10 = 99996.3  
N01 = Fi01*0.1
N1 = Fi1*1
N10 = Fi10*10
urN01 = 1/N01
urN1 = 1/N1
urN10 = 1/N10
uFi01= k95*(urTbt + urN01)*Fi01/mt.sqrt(3)
uFi1= k95*(urTbt + urN1)*Fi1/mt.sqrt(3)
uFi10= k95*(urTbt + urN10)*Fi10/mt.sqrt(3)

print(f'Incertidumbre de Frecuencia a 100Khz indicada en Tc = 0.1s: {uFi01}hz\n')
print(f'Incertidumbre de Frecuencia a 100Khz indicada en Tc = 1s: {uFi1}hz\n')
print(f'Incertidumbre de Frecuencia a 100Khz indicada en Tc = 10s: {uFi10}hz')
```
